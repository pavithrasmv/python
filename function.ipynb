```python
def func(key, value):
  return {key: value}

d1=func("a",1)
d2=func("a",2)
print(f"dictionary: {d1}, {d2}")
```

```python
def default_fix(x,lst=None):
  if lst is None:
    lst=[]
    lst.append(x)
    return lst
d1=default_fix(1,lst=None)
d2=default_fix(2,lst=None)
print(f"{d1}")
print(f"{d2}")

```

```python
func=[]
for i in range(3):
  def fun():
    print(i)

  func.append(fun)
for funct in func:
  funct()
```

```python
func=[]
for i in range(3):
  def fun(i=i):
    print(i)

  func.append(fun)
for funct in func:
  funct()
```

```python
a=50  #global variable
def local():
  b=60 # local variable
  print("value of b:",b)

print("value of a:",a)

local()

```

```python
x=50  # global variable
def local():
  global x
  x=100
  print("now global variable value change inside function:",x)
local()
print("global variable value changes outside the function after defining inside the function:",x)
```
