```python
def default_fix(x,lst=None):
  if lst is None:
    lst=[]
    lst.append(x)
    return lst
d1=default_fix(1,lst=None)
d2=default_fix(2,lst=None)
print(f"{d1}")
print(f"{d2}")

```

```python
x=50  # global variable
def local():
  global x
  x=100
  print("now global variable value change inside function:",x)
local()
print("global variable value changes outside the function after defining inside the function:",x)
```

```python
# keyword unpacking-done with dictionary
def dict(a,b):
  return a+b

res={"a":10,"b":20}
dict(**res)
```

```python
#args unpacking-done with tuple/list
def sum(a,b):
  return a+b
add=[10,20]
sum(*add)
```

```python
def sum(*args):
  total=0
  for i in args:
    total+=i
    print("sum:",total)
sum(*[3,5,6,7])
```

```python
def dic(**kwargs):
  return kwargs
dic(**{"a":1,"b":2})
```
