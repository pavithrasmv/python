```python
def func(key, value):
  return {key: value}

d1=func("a",1)
d2=func("a",2)
print(f"dictionary: {d1}, {d2}")
```

```python
def default_fix(x,lst=None):
  if lst is None:
    lst=[]
    lst.append(x)
    return lst
d1=default_fix(1,lst=None)
d2=default_fix(2,lst=None)
print(f"{d1}")
print(f"{d2}")

```
