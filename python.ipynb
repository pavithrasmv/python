```python
word=input("enter the word:")
reverse_word=word[::-1]
if word==reverse_word:
  print("it is a palindrome")
else:
  print("word is not a palindrome")
```

```python
text = input("Enter a string: ")

reverse = ""

for ch in text:
    reverse = ch + reverse

print("Reversed string:", reverse)
```
