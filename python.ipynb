```python
text = "apple mango apple orange mango apple"

words = text.split()

frequency = {}

for word in words:
    if word in frequency:
        frequency[word] += 1
    else:
        frequency[word] = 1

print(frequency)


```

```python
num = int(input("Enter number: "))

temp = num
power = len(str(num))
total = 0

while temp > 0:
    digit = temp % 10
    total = total + digit ** power
    temp = temp // 10

if total == num:
    print("Armstrong number")
else:
    print("Not Armstrong number")
```
