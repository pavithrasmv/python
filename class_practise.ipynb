```python
class Student:
  def __init__(self,name,age ,marks):
     self.name=name
     self.age=age
     self.marks=marks
  def display(self):
    print(f"name :{self.name},age:{self.age},marks:{self.marks}")

s1=Student("pavi",12,450)
s2=Student("moni",12,300)
s1.display()
s2.display()
```

```python
class Calculator:
  def __init__(self,a,b):
    self.a=a
    self.b=b
  def add(self):

    self.c=self.a+self.b
    print(f"addition of two numbers is {self.c}")
  def sub(self):
    self.c=self.a-self.b
    print(f"subtraction of two numbers is {self.c}")
  def multiply(self):
    self.c=self.a*self.b
    print(f"multiplication of two number is {self.c}")
  def div(self):
    self.c=self.a/self.b
    print(f"division of two numbers is {self.c}")
c1=Calculator(5,3)
c1.add()
c1.sub()
c1.multiply()
c1.div()
```

```python
class Bank_Account:
  def __init__(self,name,amount):
    self.name=name
    self.amount=amount
    print(f"welcome {self.name},your account balance is :{self.amount}")
  def deposit(self,dep):
    self.dep=dep
    self.amount+=self.dep
    print(f"you deposit amount is {self.dep},your current balance is  {self.amount}")
  def withdraw(self,withdraw):
    self.withdraw=withdraw
    self.amount-=self.withdraw
    print(f"your withdrawal amount is :{self.withdraw},your balance is {self.amount}")
  def checkbalance(self):
    print(f"your current balance is {self.amount}")

b1=Bank_Account("pavi",50000)
b1.deposit(20000)
b1.withdraw(15000)
b1.checkbalance()

```

```python
class Car:
  def __init__(self,brand,model,year):
    self.brand=brand
    self.model=model
    self.year=year
  def status(self):
    print("car started")

c1=Car("Toyota","fast",2020)
c1.status()
```

```python
class Employee:
  def __init__(self,employee_name,basic_salary):
    self.employee_name=employee_name
    self.basic_salary=basic_salary
  def calculate_bonus(self):

    self.bonus=self.basic_salary*0.10
    print(f"your salary bonus is {self.bonus}")

e1=Employee("pavi",50000)
e1.calculate_bonus()
```

```python
class Circle:
  def __init__(self,radius):
    self.radius=radius
  def area(self):
    self.area=3.14159*self.radius*self.radius
    print(f"area of a circle is {self.area}")
  def circumference(self):
    self.circumference=2*3.14159*self.radius
    print(f"circumference of a circle is {self.circumference}")

c1=Circle(3)
c1.area()
c1.circumference()

```

```python
class Book:
  def __init__(self,title,author,price):
    self.title=title
    self.author=author
    self.price=price
  def discount(self):
    self.discount=self.price-0.20
    print(f"actual price is {self.price} after discount {self.discount}")
b1=Book("dfsfs","sfdsavdf",1500)
b2=Book("2ytfyg","hsgdvsg",2000)
b1.discount()
b2.discount()
```
