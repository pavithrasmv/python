```python
class Employee:#parent class
  def __init__(self,name,salary):
    self.name=name
    self.salary=salary
  def cal_salary(self):
    print(f"the salary is {self.salary}")

class Manager(Employee):#child class inherits from parent class
  def __init__(self,name,salary,bonus):
    super().__init__(name,salary)
    self.bonus=bonus
  def cal_salary(self):
    calculat_salary=self.salary+self.bonus
    print("the salary is :",calculat_salary)

class Developer(Employee):#another child class inherits from parent class
  def __init__(self,name,salary,overtime_pay):
    super().__init__(name,salary)
    self.overtime_pay=overtime_pay
  def cal_salary(self):
    self.salary=self.salary+self.overtime_pay
    print("the current salary for developer is :",self.salary)

m=Manager("ram",20000,1000)
d=Developer("ravi",250000,2000)
m.cal_salary()
d.cal_salary()

```
