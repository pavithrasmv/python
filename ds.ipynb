```python
list1 = [10, 20, 30]
list2 = [40, 20, 60]

merged = list1 + list2

unique = []

for i in merged:
    if i not in unique:
        unique.append(i)

print(unique)
```

```python
nested = [1, [2, 3], 4, [5, 6]]
flat_list=[]
for i in nested:
  if type(i)==list:
    for j in i:
      flat_list.append(j)
  else:
      flat_list.append(i)
print(flat_list)
```
