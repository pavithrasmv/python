```python
numbers = [1, 2, 3, 2, 4, 1]

unique = []

for i in numbers:
    if i not in unique:
        unique.append(i)

print(unique)
```

```python
list1 = [10, 20, 30]
list2 = [40, 20, 60]

merged = list1 + list2

unique = []

for i in merged:
    if i not in unique:
        unique.append(i)

print(unique)
```
